Print out bar chart comparing results from all final validation test runs

```python
# Get imports
import numpy as np  
import matplotlib.pyplot as plt  
```

```python
# Store all values and display to bar graph
X = ['Accuracy','F1-Score','TPR','FPR'] 
MLP = [88.808, 77.45,54.2,4.37] 
LGR = [86.375, 67.56,31.03,2.71] 
RFC = [89.132, 78.40, 56.65, 4.47]
GDBC = [89.213, 77.70, 52.71, 3.59]

# Set size requirements for plot visualization
X_axis = np.arange(len(X)) 
width = 0.17

# Display all bars
plt.bar(X_axis + width, MLP, 0.17, label = 'MLP') 
plt.bar(X_axis + width*2, LGR, 0.17, label = 'LGR')
plt.bar(X_axis + width*3, RFC, 0.17, label = 'RFC') 
plt.bar(X_axis + width*4, GDBC, 0.17, label = 'GDBC') 

# Put legend, labels, and edit the axes
plt.xticks(X_axis + 0.415, X) 
plt.xlabel("Evaluation Metrics") 
plt.ylabel("Amount (Some Scaled Accordingly)") 
plt.title("Validation Test Evaluation Comparisons") 
plt.ylim(0, 93)
plt.legend() 
plt.show() 
```
